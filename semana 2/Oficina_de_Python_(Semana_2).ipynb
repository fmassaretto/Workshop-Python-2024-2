# Desafio

Criar um sistema simples de cadastro de vaga de emprego utilizando o que vimos até o momento. Crie variáveis para os seguintes dados:


*   Nome
*   Email
*   Idade
*   Primeiro emprego (Booleano: True ou False)
*   Pretenção Salarial

Obs.: Podem guadar esses dados como quiserem: em variaveis simples, ou dicionario

Imprima os dados, ex:

Nome: John

Email: john@gmail.com

Idade: 18

Primeiro emprego? True (Se ja tiver algum conhecimento poderá tratar esse dado e mostrar sim ou não, em vez de True ou False)

Pretenção Salarial: 1500.00

# Resolução

```python
# Existem varias formas de fazer, essa é uma delas

nome = input("Digite seu nome: ")
email = input("Digite seu email: ")
idade = int(input("Qual a sua idade? "))
primeiro_emprego = bool(input("É seu primeiro emprego? (Digite True para sim ou False para nao)"))
# Tem formas melhores de fazer a linha acima, é mais para treinar o uso da função bool(), entao vc poderia somente digitar sim ou nao em string mesmo, ex:
primeiro_emprego_2 = input("É seu primeiro emprego? (Digite sim ou nao)") # segunda forma de fazer
pretencao_salarial = float(input("Qual a sua pretenção salarial? "))


# Uma forma mais completa e usando if-else (será demonstrado na proxima oficina)
# verifico se primeiro_emprego é igaul a True, se sim, guardo um texto em uma nova variavel
# se nao, guardo outro texto
if primeiro_emprego == True:
  primeiro_emprego_longo = "Sim, nunca trabalhei antes"
else:
  primeiro_emprego_longo = "Não, já trabalhei antes"

# Com variaveis simples e f-string
print(f'Nome: {nome}')
print(f'Email: {email}')
print(f'Idade: {idade}')
print(f'É seu primeiro emprego? {primeiro_emprego_longo}')
print(f'É seu primeiro emprego? {primeiro_emprego_2}') # segunda forma de fazer
print(f'Pretenção salarial: {pretencao_salarial}')
```

# Extra

Curiosidades do f-string

```python
# 1 - separador de milhares com _ (underscore) ou , (virgula)
n: int = 100000000

print(f'{n:_}')
print(f'{n:,}')

# 2 - colocar espaços em brando na esquerda, na direita ou centro
var: str = 'variavel'

print(f'{var:>20}:')
print(f'{var:$<20}:') # ou também, pode preencher os espaços com algum caracter '$' neste caso
print(f'{var:<20}:')
print(f'{var:^20}:')

# 3 - Arredondar casas decimais
f: float = 23123.4164124

print(f'{f:.2f}')
print(f'{f:.0f}') # sem casas decimais
print(f'{f:,.2f}') # com separador de milhares

# 4 - Fazer operações direto no f-string
a: int = 5
b: int = 10

print(f'a + b = {a + b}') # jeito correto porém mais verboso
print(f'{a + b = }') # jeito mais enxuto
```
