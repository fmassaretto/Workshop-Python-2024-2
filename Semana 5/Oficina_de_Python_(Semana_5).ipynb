# Imports
A instrução "Import" em Python é usada para acessar e utilizar código definido em outros arquivos ou bibliotecas. Ela permite que você inclua funcionalidades em seus softwares.

Basicamente, existem duas formas de importar uma biblioteca/modulo para o programa
1. import modulo
2. from modulo import função

Uma observação importante é que em Python, tanto o 1 quanto 2, importam todo o conteúdo do modulo, o 2 jeito ele só facilita para gente chamar a função diretamente (ver os exemplos para entender melhor)

Se quiser pode dar um "apelido" para o modulo, para que não precise escrever toda vez o nome do modulo, ex.:
1. import pandas as pd

Se o Python não encontrar o modulo, ele poderá dar o seguinte erro:


```
ModuleNotFoundError: No module named 'pandas'
```
Provavelmente ou o nome do modulo está errado (verificar graáfia), ou precisa instalar o modulo. Para instalar deve abrir um terminal/console e digitar o comando:


```
pip install nome_do_modulo
```

# Função

As funções permitem definir um bloco de código reutilizável que pode ser executado muitas vezes dentro de um programa. É uma estrutura que agrupa partes do código, criando soluções modulares para problemas complexos.

Para definir uma função, escreve-se assim:


```
def nome_da_funcao():
  # seu código aqui
```

E em algum momento quando eu quiser chamar/executar a função é só escrever o nome da função com parênteses no final, assim:


```
nome_da_funcao()
```

Uma função pode não ter nenhum retorno, ou seja, quando eu chamar a função ela não irá devolver nenhum valor, ex.:


```
def funcao_sem_retorno:
  print('Não retorna nenhum valor')
```

Ou pode devolver algum valor, e quando queremos devolver algum valor devemos adicionara a palavra return na função, ex.:


```
def funcao_com_retorno():
  return 'Aqui eu posso devolver qualquer tipo de dados, que já aprendemos'
```

Podemos passar parâmetros para a função, adicionando variáveis dentro dos parênteses, separados por vírgula se tiver mais que um parâmetro, ex.:


```
def funcao_somar_com_parametros(x, y);
  return x + y
```
