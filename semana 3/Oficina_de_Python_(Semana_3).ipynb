A estrutura basica do if-elif-else:

```
# This is formatted as code
if(Condicao):
  # seu codigo aqui se a condicao for verdadeira
elif(Condicao):
  # seu codigo aqui se essa condicao for verdadeira
else:
  # seu codigo aqui se todas as outras condicoes NAO forem verdadeira
```



````python
operacao_aritmetica = 'modulo'
x = 10
y = 20

if(operacao_aritmetica == 'subtracao' or operacao_aritmetica == 'adicao'):
  print(f'subtracao {y - x = }')
  print(f'adicao {x + y = }')
elif(operacao_aritmetica == 'divisao' or operacao_aritmetica == 'multiplicacao'):
  print(f'divisao {x / y = }')
  print(f'multiplicacao {x * y = }')
elif(operacao_aritmetica == 'potenciacao'):
  print(f'potenciacao {x ** y = }')
elif(operacao_aritmetica == 'modulo'):
  print(f'Modulo (resto) {x % 3 = }') # muito usado para saber se um numero é par ou impar
else:
  print('operacao invalida')
````

````python
hora_do_dia = 6

if(hora_do_dia >= 4 and hora_do_dia < 12):
  print('bom dia')
elif (hora_do_dia >= 12 and hora_do_dia < 18):
  print('boa tarde')
elif (hora_do_dia >= 18 and hora_do_dia < 24):
  print('boa noite')
else:
  print('boa madrugada')
````

# Desafio 1

Calcular IMC, pedindo a altura e o peso do usuario. Deixei um pdf no teams e no github(https://github.com/fmassaretto/Workshop-Python-2024-2)

````python
altura = float(input('Sua altura: '))
peso = float(input('Seu peso: '))

imc = peso / (altura * altura) # ou outra forma: imc = peso / (altura ** 2)
# obs.: o Parêntese da prioridade em resolver o altura ao quadrado primeiro e depois divide com o peso

if(imc < 18.5):
    print('Abaixo do peso, imc =', imc)
elif(18.5 <= imc < 24.9):
    print('Peso ideal, imc =', imc)
elif(25.0 <= imc < 29.9):
    print('Excesso peso, imc =', imc)
elif(30.0 <= imc < 34.9):
    print('Obs classe 1, imc =', imc)
elif(35.0 <= imc < 39.9):
    print('Obs classe 2, imc =', imc)
else:
    print('Obs classe 3, imc =', imc)

````
